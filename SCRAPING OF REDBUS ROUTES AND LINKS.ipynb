# KERALA ROUTES

```python
driver = webdriver.Edge()

driver.get("https://www.redbus.in/online-booking/ksrtc-kerala/?utm_source=rtchometile")

time.sleep(3)

driver.maximize_window()


from selenium.common.exceptions import (
    NoSuchElementException,
    TimeoutException,
    ElementClickInterceptedException,
)
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
wait = WebDriverWait(driver,10)

def kerala_link_route(path):
    links_kerala = []
    routes_kerala = []
    

    for i in range(1,6):
        try:
            # Retrieve route links and names
            paths = driver.find_elements(By.XPATH, path)
            for link in paths:
                kl = link.get_attribute("href")
                links_kerala.append(kl)
                routes_kerala.append(link.text)

            # Pagination handling
            pagination_xpath =  '//*[@class="DC_117_paginationTable"]'
            next_button_xpath = f'//div[@class="DC_117_pageTabs " and text()={i + 1}]'
            
            # Wait for pagination and next button
            pagination = wait.until(EC.presence_of_element_located((By.XPATH, pagination_xpath)))
            next_button = pagination.find_element(By.XPATH, next_button_xpath)
            time.sleep(3)
            next_button.click()
            
        except ElementClickInterceptedException:
            print("Next button click intercepted. Retrying...")
            driver.execute_script("arguments[0].click();", next_button)
        except TimeoutException:
            print("No more pages to paginate.")
            break
        except NoSuchElementException:
            print(f"No next button found for page {i}.")
            break

    return links_kerala, routes_kerala

# Call the function
links_kerala, routes_kerala = kerala_link_route("//a[@class='route']")
```

```python
df_k=pd.DataFrame({"Route_name":routes_kerala,"Route_link":links_kerala})
df_k
```

```python
# change dataframe to csv
path="D:/REDBUS PROJECT/df_k.csv"
df_k.to_csv(path,index=False)
```
